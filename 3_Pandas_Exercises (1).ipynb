1. Import the necessary libraries

```python
import pandas as pd
```

```python
import numpy as np
```

2. Import the dataset iris.data

3. Assign it to a variable called iris

```python
iris = pd.read_csv("iris.data")
```

4. Create columns for the dataset

```python
# 1. sepal_length (in cm)
# 2. sepal_width (in cm)
# 3. petal_length (in cm)
# 4. petal_width (in cm)
# 5. class
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
```

```python
iris
```

5. Is there any missing value in the dataframe?

```python
iris.isnull().sum()
```

6. Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```python
iris['petal_length'].iloc[10:30] = np.NaN
```

```python
iris.head(30)
```

7. Good, now lets substitute the NaN values to 1.0

```python
iris['petal_length'].fillna(1.0, inplace=True)
```

```python
iris.head(30)
```

8. Now let's delete the column class

```python
iris.drop('class', axis = 1, inplace=True)
```

```python
iris
```

9. Set the first 3 rows as NaN

```python
iris.iloc[0:3] = np.NaN
```

```python
iris
```

10. Delete the rows that have NaN

```python
iris.dropna(inplace=True)
```

```python
iris.reset_index(drop=True, inplace=True)
```
