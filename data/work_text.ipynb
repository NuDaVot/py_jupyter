```python
def preprocess_text(data, stopwords=stopwords.words("russian")):
    stopwords.extend(['либо', 'это', 'мб', 'далее', 'дв', 'свой', 'ваш','всё', 'очень', 'её', 'ещё', 'вообще', 'наш', 'который',])
    text = re.sub('ё','е', data.lower())
    text = re.sub(r"([.,!?])", r" \1 ",text)
    text = re.sub(r"[^а-яА-Я\s]+", "",text)
    text = text.strip()
    text = [w for w in text.split() if w not in stopwords] 
    text = [w for w in text if len(w) >=3]
    return ' '.join(text)
```

```python
def edit_text(text):
    
    text = ''.join([i if i.isalpha() else ' ' for i in text])
    text = ' '.join([i if len(i) >= 3 else ' ' for i in text.split(" ")])
    text = text.lower()
    text = text.replace('ё', 'е')
    text = text.replace('й', 'и')   
    text = re.sub('[a-z]',' ', text, flags=re.I)
    text = ' '.join([word  if word not in stopword else ' ' for  word in text.split(" ")])
    text = re.sub(r'\s+',' ', text, flags=re.I)
    return text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.8, max_features=100000,
                                 min_df=0.01, stop_words=stopwords.words("russian"),
                                 ngram_range=(1,3))
```

```python
%%time
tfidf_matrix = tfidf_vectorizer.fit_transform(df['prep_text3'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
RAND = 10
lda_model = LatentDirichletAllocation(learning_method='online', random_state=RAND, n_jobs=-1, n_components=7)
```

```python
lda_model.fit(tfidf_matrix)
```

```python
import pyLDAvis.sklearn
```

```python
panel = pyLDAvis.sklearn.prepare(lda_model, tfidf_matrix, tfidf_vectorizer, mds='tsne')
panel
```

```python
for i,topic in enumerate(lda_model.components_):
    print(f'Top 10 words for topic #{i + 1}:')
    print([tfidf_vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print('\n')
```
