```python
def pars_pdf(text):
    text2 = text.split('\n')
    if 'Рейтинг' in text2:
        num = text2.index('Рейтинг')
        raiting = text2[num-1]
        name = text2[num+1]
        activity = text2[num+2]
        
        for i in range(len(text2)):
            if "назад" in text2[i]:
                a = re.findall('(\d+)', text2[i])
                date_publish = text2[i][text2[i].rfind(str(a[len(a)-1])):]
    else:
        return 
#         raiting = 0
#         name = ""
#         activity = ""
#         date_publish = ""
#         if 'Простой' in text2:
#             num2 = text2.index('Простой')
#             date_publish = text2[num2-1]
#         text2=""
#         return name, raiting, activity, date_publish, text
   
    if date_publish == None:
        return  name, raiting, activity 
    else:
        text2.remove('Рейтинг')
        text2.remove(raiting)
        text2.remove(name)
        text2.remove(activity)
        text2.pop(0)
        if("Оценка" in text2):
            text2.remove('Оценка')
        for i in text2:
            if "Средний" in i:
                text2.remove(i)
            if "Автор оригинала" in i:
                text2.remove(i)
           
        if name == "red_mad_robot":
            text2 = ""
        return  name, raiting, activity, date_publish,text2
```
