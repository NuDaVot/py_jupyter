```python
import requests
from bs4 import BeautifulSoup

# Получаем HTML-код веб-страницы
url = "https://example.com"
response = requests.get(url)
html_code = response.content

# Парсим HTML-код и извлекаем информацию
soup = BeautifulSoup(html_code, 'html.parser')
title = soup.title.text
paragraphs = soup.find_all('p')

# Выводим полученную информацию
print("Заголовок страницы:", title)
print("Параграфы на странице:")
for p in paragraphs:
    print(p.text)
```

```python
import PyPDF2

# Открываем PDF-файл и создаем объект для чтения данных
with open('example.pdf', 'rb') as pdf_file:
    reader = PyPDF2.PdfFileReader(pdf_file)

    # Читаем страницы PDF-файла и извлекаем текст
    for page in range(reader.numPages):
        pdf_page = reader.getPage(page)
        page_text = pdf_page.extractText()

        # Выводим извлеченный текст на экран
        print("Текст на странице", page+1, ":")
        print(page_text)
```
