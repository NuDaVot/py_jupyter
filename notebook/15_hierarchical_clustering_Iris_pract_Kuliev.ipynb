Загрузка Библеотек 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import datasets 
plt.rcParams['figure.figsize'] = (15,5)
```

Загразка файла 

```python
data = datasets.load_iris()
data
```

Загрузка данных 

```python
iris = data["data"]
iris
```

Создание Data Ftame

```python
df = pd.DataFrame(iris, columns = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'])
df
```

Иерархическая кластеризация от scipy

```python
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram 
```

```python
Z = linkage(iris, method='average', metric='euclidean')
```

```python
dend = dendrogram(Z, color_threshold=0.0)
```

```python
label = fcluster(Z, 1.9, criterion='distance')
```

```python
np.unique(label)
```

Колическо классетров 3 

```python
df["Prediction_scipy"] = label
```

```python
for i, group in df.groupby('Prediction_scipy'):
    print('=' * 10)
    print('cluster {}'.format(i))
    print(group)
```

Иерархическая кластеризация от sklearn

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
label = hc.fit_predict(iris)
```

```python
df["Prediction_sklearn"] = label
```

```python
df
```

Таблица рузультатов 

```python
d = {'Target' : df.target, 'Prediction_scipy' : df.Prediction_scipy, 'Prediction_sklearn' : df.Prediction_sklearn}
 
# превратим словарь в датафрейм с соответствующими названиями столбцов
df_rez = pd.DataFrame(d, columns = ['Target', 'Prediction_scipy', 'Prediction_sklearn'])
df_rez
```

```python
comparison_scipy = np.where(df_rez['Target'] == df['Prediction_scipy'], True, False)
```

```python
comparison_sklearn = np.where(df_rez['Target'] == df['Prediction_sklearn'], True, False)
```

```python
result = df_rez
result['Prediction_scipy'] = pd.DataFrame(comparison_scipy)
result.head()
```

```python
result['Prediction_scipy'].value_counts(normalize = True).round(2)
```

Предсказания от scipy показало 91% верных значений 

```python
result = df_rez
result['Prediction_sklearn'] = pd.DataFrame(comparison_sklearn)
result.head()
```

```python
result['Prediction_sklearn'].value_counts(normalize = True).round(2)
```

Предсказания от scipy показало 89% верных значений
