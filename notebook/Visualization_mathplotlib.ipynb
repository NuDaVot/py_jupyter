
### Построение графика

#### Functional method

```python
# Построение графика
plt.title('Plot for x and y1 arrays') # заголовок
plt.xlabel('x') # ось абсцисс
plt.ylabel('y1') # ось ординат
plt.grid() # включение отображения сетки
plt.plot(x, y1) # построение графикa
```

```python
# Построение графика
plt.title('Plot for x and y1 arrays') # заголовок
plt.xlabel('x') # ось абсцисс
plt.ylabel('y1') # ось ординат
plt.grid() # включение отображения сетки
plt.plot(x, y1, 'r--', label = 'y1 = x ^ 3') # построение графикa
plt.legend()
```

### Построение несколько графиков на одном поле

```python
y2 = x
```

```python
# Построение графика
plt.title('Зависимости: y1 = x^3, y2 = x') # заголовок
plt.xlabel('x') # ось абсцисс
plt.ylabel('y1, y2') # ось ординат
plt.grid() # включение отображения сетки
plt.plot(x, y1, x, y2) # построение графика
plt.legend(['y1 = x^3', 'y2 = x'])
```

### Представление графиков на разных полях

```python
# Построение графиков
plt.figure(figsize=(9, 9))

plt.subplot(2, 1, 1)
plt.plot(x, y1) # построение графика
plt.title('Зависимости: y1 = x^3, y2 = x') # заголовок
plt.ylabel('y1', fontsize=14) # ось ординат
plt.grid(True) # включение отображение сетки

plt.subplot(2, 1, 2)
plt.plot(x, y2) # построение графика
plt.xlabel('x', fontsize=14) # ось абсцисс
plt.ylabel('y2', fontsize=14) # ось ординат
plt.grid(True) # включение отображение сетк
```

figure() — функция для задания глобальных параметров отображения графиков. В неё через параметр figsize передаётся кортеж из двух элементов, определяющий общий размер фигуры.

subplot() — функция для задания местоположения поля с графиком. Существует несколько способов задания областей для
вывода графиков. Первый аргумент — количество строк, второй — столбцов в формируемом поле, третий — индекс (номер поля, считаем сверху вниз, слева направо).
