# Описание признаков

Pregnancies - Количество беременностей

Glucose - Концентрация глюкозы в плазме крови через 2 часа при проведении перорального теста на толерантность к глюкозе

BloodPressure - Диастолическое артериальное давление (мм рт. ст.)

SkinThickness - Толщина кожной складки трицепса (мм)

Insulin - 2-часовой сывороточный инсулин (мю Ед/мл)

BMI - Индекс массы тела (вес в кг/(рост в м)^2)

DiabetesPedigreeFunction - Функция составления родословной диабета

Age - Возраст (годы)

Outcome - Переменная класса (0 или 1) 268 из 768 - 1, остальные - 0

```python
column = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
for i in column:
    print(diabetes[diabetes[i] == 0][i].value_counts())
    
```

```python
column = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
diabetes1 = diabetes.copy(deep=True)
diabetes1[column] = diabetes1[["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]].replace(0, np.nan)
diabetes1.isnull().sum()
```

```python
for i in column:
    diabetes1[i].fillna(diabetes1[i].mean(), inplace=True)
    
diabetes1.info()
```

Сосчитали количество нулевых значений у признаков Glucose, BloodPressure, SkinThickness, Insulin, BMI и заменили эти данные соотвестующими средними значениями

# Разбиение выборки на тестовую и обучающую

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
X = diabetes1.iloc[:, :8]
y = diabetes1["Outcome"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=3, stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.value_counts())
```

```python
print(y_test.value_counts())
```

```python
Scaler = StandardScaler()
X_train = Scaler.fit_transform(X_train)
```

```python
X_test = Scaler.transform(X_test)
```

```python
X_train
```

Разбили данные на обучающие и тестовые в соотношении 1:3 страфицированно. 

Проделали масштабирование данных

## Метод близайших соседей

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import rand_score

from sklearn.metrics import classification_report

pred_train = knn.predict(X_train)
pred_test = knn.predict(X_test)

print(f'Train knn {rand_score(y_train, pred_train):.2f}')
print(f'Test knn {rand_score(y_test, pred_test):.2f}')
```

```python

Pred = knn.predict(X_test)
print(classification_report(y_test, Pred))
```

Точность модели на тренировочных данных составляет 72%, на тестовых - 62%


# Метод классификаторов дерева решений

```python
from sklearn.tree import DecisionTreeRegressor

tree = DecisionTreeRegressor(random_state=1)
```

```python
tree.fit(X_train, y_train)
```

```python
pred_train = tree.predict(X_train)
pred_test = tree.predict(X_test)

print(f'Train tree {rand_score(y_train, pred_train):.2f}')
print(f'Test tree {rand_score(y_test, pred_test):.2f}')
```

```python

Pred = tree.predict(X_test)
print(classification_report(y_test, Pred))
```

Модель переобучилась, поскольку на тренировочных данных 100% правильных предсказаний, возможно нажно поиграсться с глубиной дерева

```python
tree_6 = DecisionTreeRegressor(random_state=1 , max_depth=6)
tree_6.fit(X_train, y_train)
```

```python
pred_train = tree_6.predict(X_train)
pred_test = tree_6.predict(X_test)

print(f'Train tree {rand_score(y_train, pred_train):.2f}')
print(f'Test tree {rand_score(y_test, pred_test):.2f}')
```

Показатели улучшелись 
Точность модели на тренировочных данных составляет 59%, на тестовых - 12%

## Метод опорных векторов 

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
gnb
```

```python
pred_train = gnb.predict(X_train)
pred_test = gnb.predict(X_test)

print(f'Train gnb {rand_score(y_train, pred_train):.2f}')
print(f'Test gnb {rand_score(y_test, pred_test):.2f}')
```

Точность модели на тренировочных данных составляет 62%, на тестовых - 65%, что являеться наилучшим резудьтатом 

## Подберем наилучшие параметры для наилучшей  модели, а именно Метод опорных векторов

```python
from sklearn.model_selection import GridSearchCV
parametrs = { 'priors': ["Pregnancies", 2, 3, 4, 5, 6, 7, 8, 9, None]}
grid = GridSearchCV(gnb, parametrs, cv=5)
```

```python
%%timeit

# обучение 
grid.fit(X_train, y_train) 
```

```python
# параметры лучшей модели
grid.best_params_ 
```

```python
# лучший алгоритм
best_knn = grid.best_estimator_ 
```

```python
pred_train = best_knn.predict(X_train)
pred_test = best_knn.predict(X_test)

print(f'Train best_knn {rand_score(y_train, pred_train):.2f}')
print(f'Test best_knn {rand_score(y_test, pred_test):.2f}')
```

Подобрав науличший параметр мы пришли к точности в 65% на тестовых данных
