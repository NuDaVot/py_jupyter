```python
#Подключаем библиотеки
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import docx
import pdfkit
import io
import types
```

```python
#GET-запрос
url = 'https://habr.com/ru/all'
page = requests.get(url)
```

```python
soup = bs(page.text, 'html.parser')
```

```python
result_list = {'title': [], 'namecompany': [], 'description': [], 'rating': [], 'field': [], 'date': []}
```

```python
pagenum = 1 #переменная для счета страниц
for i in range(10):
    url = 'https://habr.com/ru/all/page' + str(pagenum) + '/' # переход на ссылку с определенным номером страницы
    page = requests.get(url)
    soup = bs(page.text, 'html.parser')
    pages = soup.find_all('h2', class_= 'tm-article-snippet__title') # получение заголовков статей
    for i in pages: #перебор статей
        fieldtext = "" #отрасли компании
        url = 'https://habr.com' + str(i.a.get('href')) #переход на страницу статьи
        page = requests.get(url)
        soup = bs(page.text, 'html.parser')
        name_company = soup.find('a', class_= 'tm-company-snippet__title') #получаем название компании
        desc_company = soup.find('div', class_='tm-company-snippet__description') #описание компании
        if(name_company is not None): #если на странице статьи присутствует название компании
            result_list['title'].append(i.text) #заполняем словарь названием статьи
            result_list['namecompany'].append(name_company.text)#заполняем словарь названием компании
            result_list['description'].append(desc_company.text)#заполняем словарь описанием компании
            datepub = soup.find('span', class_= 'tm-article-datetime-published')#получаем дату публикации
            result_list['date'].append(datepub.time['datetime'])#записываем дату публикации в словарь
            url = 'https://habr.com' + str(name_company.get('href'))#переходим на страницу компании
            page = requests.get(url)
            soup = bs(page.text, 'html.parser')
            #получаем рейтинг компании
            rating = soup.find('span', class_='tm-votes-lever__score-counter tm-votes-lever__score-counter tm-votes-lever__score-counter_rating')
            if(rating is None): #если рейтинг отсутствует то добавляем в словарь 0
                result_list['rating'].append('0')
            else:
                result_list['rating'].append((rating.text).strip())
            field = soup.find_all('a', class_='tm-company-profile__categories-text') #получаем список отраслей компании
            for fields in field:
                fieldtext = fieldtext + ((fields.text).strip()) + " "
            fieldtext = fieldtext.splitlines()
            result_list['field'].append(''.join(str(e) for e in fieldtext))# добавляем отрасли в словарь
    pagenum = pagenum + 1 # прибавляем страницу
    print(result_list)
```

```python
FILE_NAME = "habr.csv"
df = pd.DataFrame(data=result_list)
df.to_csv(FILE_NAME)
```

```python
content = docx.Document('companies.docx').paragraphs
```

```python
companies = []
for i in range(len(content)):
    companies.append(content[i].text)
companies
```

```python
#GET-запрос
url = 'https://habr.com/ru/all'
page = requests.get(url)
```

```python
soup = bs(page.text, 'html.parser')
```

```python
result_list = {'title': [], 'namecompany': [], 'description': [], 'rating': [], 'field': [], 'date': []}
```

```python
pagenum = 1
for i in range(50):
    url = 'https://habr.com/ru/all/page' + str(pagenum) + '/'
    page = requests.get(url)
    soup = bs(page.text, 'html.parser')
    pages = soup.find_all('h2', class_= 'tm-article-snippet__title')
    for i in pages:
        fieldtext = ""
        url = 'https://habr.com' + str(i.a.get('href'))
        page = requests.get(url)
        soup = bs(page.text, 'html.parser')
        name_company = soup.find('a', class_= 'tm-company-snippet__title')
        desc_company = soup.find('div', class_='tm-company-snippet__description')
        if(name_company is not None and name_company.text in companies):
            result_list['title'].append(i.text)
            result_list['namecompany'].append(name_company.text)
            result_list['description'].append(desc_company.text)
            datepub = soup.find('span', class_= 'tm-article-datetime-published')
            result_list['date'].append(datepub.time['datetime'])
            url = 'https://habr.com' + str(name_company.get('href'))
            page = requests.get(url)
            soup = bs(page.text, 'html.parser')
            rating = soup.find('span', class_='tm-votes-lever__score-counter tm-votes-lever__score-counter tm-votes-lever__score-counter_rating')
            if(rating is None):
                result_list['rating'].append('0')
            else:
                result_list['rating'].append(rating.text)
            field = soup.find_all('a', class_='tm-company-profile__categories-text')
            for fields in field:
                fieldtext = fieldtext + fields.text
            result_list['field'].append(fieldtext)
            print("Название статьи: " + i.text + "Компания: " + name_company.text + "Описние компании: " + desc_company.text + "Дата: " + datepub.time['datetime'])
    pagenum = pagenum + 1
```

```python
FILE_NAME = "habr.csv"
df = pd.DataFrame(data=result_list)
df.to_csv(FILE_NAME)
```
